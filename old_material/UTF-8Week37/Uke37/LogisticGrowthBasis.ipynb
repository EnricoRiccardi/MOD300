```python

import matplotlib.pyplot as plt
import numpy as np
import math

timestep = 1 #  Use month as timeunit.

starttime = 0
endtime = 12*15  # Number of months that shall be simulated  

K = 10000   # K = 10000 kg
r = 0.6/12  # per month

N0  = 2000 # Initial  biomass in kg

T1=[]
YExact=[]

t = starttime
y = N0


YExact.append(y)
T1.append(t)


while (t<endtime):
    t = t+timestep 
   
    
    y=N0*K/(N0+(K-N0)*math.exp(-r*t))  # Exact solution
    T1.append(t)
    YExact.append(y)
    
     
plt.plot(T1,YExact)    
plt.title('Fishpopulation') 
plt.xlabel('Time (months)')
plt.ylabel('Biomass (kg)')
plt.ylim(ymin=0,ymax =20000)
plt.xlim(xmin=0)
plt.grid(True)
plt.legend(['Exact'])
plt.show()
    



```
